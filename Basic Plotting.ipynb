```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This is a "magic" command for inline plotting within Jupyter Notebook
# https://stackoverflow.com/questions/43027980/purpose-of-matplotlib-inline
%matplotlib inline
```

```python
random_list1 = 500.*np.random.rand(500)
random_list2 = 250.*np.random.rand(500)
```

## The first way to plot

```python
# plt.figure creates a figure, and sets it as the active figure.
# While the method returns a Figure instance, we don't store it in a variable in this case.
# The only way to refer to then refer to it is via the plt method

plt.figure(figsize = (5, 5)) # Setting the size of figure

# Make a scatter plot with your data
plt.scatter(random_list1[::-1], random_list1, color = 'black', marker = 'o', s = 50)

# There is no 'overplot' call in python - data is simply plotted on the 
# currently active set of axes
# Scatter plot of another dataset
plt.scatter(random_list2[::-1], random_list2, color = 'orange', marker = 'x', s = 54)

plt.xlabel('Weight', fontsize=12)
plt.ylabel('Height', fontsize=12)

# It is curious that tickmarks and labels are different attributes of the plot
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)

# The attributes of the axes
plt.xlim(-100, 800)
plt.ylim(0, 800)

# Generally there is a final plt.show() command which displays the image
# Jupyter automatically displays the figure after generating the plot
```

```python
plt.figure(figsize = (5, 5)) # Setting the size of figure

# 
plt.scatter(random_list1[::-1], random_list1, marker = 'o', s = 50, color = 'black', label = 'black')
plt.scatter(random_list2[::-1], random_list2, color = 'r', marker = 'x', s = 54, label = 'red')

# It is curious that tickmarks and labels are different attributes of the plot

plt.xticks(fontsize=12, rotation=90)
#plt.yticks(fontsize=12, rotation=90)

plt.xlabel('weight', fontsize=12)
plt.ylabel('height', fontsize=12)

plt.legend()
```

```python
x_points1 = np.arange(-10, 10, 0.1)
parabola1 = x_points1 * x_points1

x_points2 = np.arange(-5, 15, 0.1)
parabola2 = x_points2*x_points2
```

```python
plt.figure(figsize = (5, 5))
plt.plot(x_points1, parabola1, color = 'black', label = 'black')
plt.plot(x_points2, parabola2, color = 'red', label = 'red')

plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12, rotation=45)

plt.xlabel('Some X-title', fontsize=15)
plt.ylabel('Some Y-title', fontsize=15)

plt.legend()

path = 'C:\\Users\\karth\\Desktop\\'
plt.savefig(path + 'test_image.eps')
```

## The other way

```python
# The other way to do plots is to instantiate the figure
# and save the figure and axes references in variables
# The figure and axes attributes can then be subsequently modified
# Axes attributes : https://matplotlib.org/api/axes_api.html
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (6, 3))
# This can also be done as :
# fig, ax_lst = plt.subplots(nrows = 1, ncols = 2, figsize = (6, 3))
# where ax_lst[0] and ax_lst[1] are the references to the axes

ax1.plot(x_points1, parabola1, color = 'blue')
ax1.set_title('Blue Parabola (1)')
ax1.set_xlim((-10, 20))
ax1.set_ylim((-10, 20))

ax2.plot(x_points1, parabola2, color = 'red')
ax2.set_title('Red Parabola (2)')
```

```python
# Fig and ax are the references to the figure and the axes within the plots
fig, ax = plt.subplots(figsize=(5, 5))

ax.plot(x_points1, parabola1, color = 'blue')
ax.set_title('Blue Parabola (1)', fontsize = 20)
ax.set_xlabel('X-axis', fontsize = 15)
ax.set_ylabel('Y-axis', fontsize = 15)

#ax.xticks(fontsize=12)
#ax.yticks(fontsize=12, rotation=45)
ax.tick_params(axis='x', labelsize = 25, rotation = 45)
ax.tick_params(axis='y', labelsize = 25, rotation = 90)
# Or you could do:
#ax.tick_params(axis='both', labelsize = 25, rotation = 45)
```

```python
# This is just a sample plot from the website
# http://www.pybloggers.com/2018/02/python-plotting-with-matplotlib-guide/

rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

csfont = {'fontname':'Arial'}
hfont = {'fontname':'Arial'}

fig, ax = plt.subplots(figsize=(5,5))
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')

ax.set_title('Combined debt growth over time',**hfont, fontsize=15)
ax.set_xlabel('Years', **hfont, fontsize=15)
ax.set_ylabel('Total debt', **hfont, fontsize=20)

plt.xticks(fontsize=20, rotation=90, **csfont)

ax.legend(loc='upper left')
#ax.set_ylabel('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
path = 'C:\\Users\\karth\\Desktop\\'
plt.savefig(path + 'test_image1.png')
```
